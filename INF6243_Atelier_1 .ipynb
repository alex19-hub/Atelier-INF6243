Création d'une calculatrice basique qui effectue les 4 opérations arithmétiques + , - , *, /

````python
def calc():
  chiffre_1 = int(input("Veuillez saisir le premier chiffre svp: "))
  operation = str(input("Veuillez saisir l'operation svp: "))
  chiffre_2 = int(input("Veuillez saisir le 2eme chiffre svp: "))

  clear_output(wait=True)

  if (operation == '+'):
    print(chiffre_1 + chiffre_2)


  elif (operation == '-'):
    print(chiffre_1 - chiffre_2)

  elif (operation == '*'):
    print(chiffre_1 * chiffre_2)

  elif (operation == '/'):
    print(chiffre_1 / chiffre_2)

  else:
    print("Je ne peux pas effectuer ce genre d'opérations")
````

````python
calc()
````

````python
# Définir une fonction qui calcule la distance entre 2 points a et b
import math

def distance(Xa = 0, Ya = 0, Xb = 0, Yb = 0):

  dist = math.sqrt((Xb - Xa)**2 + (Yb - Ya)**2)

  return print(f"la distance entre le point {Xa, Ya} et le point {Xb, Yb} est: ", dist)
````

The lower standard deviation and lower variance, the less spread out the values are.

# **1. Les types de variables**

En Python, pour donner une valeur à une variable, il suffit d'écrire **nom_de_la_variable = valeur**.

Les noms de variables Python peuvent contenir tous les caractères alphabétiques (AZ et az), les chiffres (0-9) et le caractère de soulignement "_". Il y a quelques mises en garde et restrictions:

* Le premier caractère de la variable doit être un alphabet ou un trait de soulignement (_).

* Un chiffre ne peut pas commencer un nom de variable. Par conséquent, étudiant1 est une variable valide, alors que le 1er étudiant ou 1etudiant ne l'est pas.

* Le nom de l'identifiant ne doit contenir aucun espace blanc ni de caractère spécial (!, @, #,%, ^, &, *).

* Le langage Python est sensible à la casse ce qui signifie que les lettres majuscules et minuscules sont distinctes. Par conséquent, les lettres majuscules et minuscules ne constituent pas la même variable . Par exemple, les variable nom, Nom, nOm, NOM, noM sont toutes différentes les unes des autres.

* Les mots-clés et noms de fonction intégrés ne doivent pas être utilisés comme nom de variable. Par exemple, print est évidemment un mot-clé Python, vous ne devriez donc pas l'utiliser comme nom de variable.

### Accéder aux valeurs dans les chaînes
Python ne prend pas en charge un type de caractère ; ceux-ci sont traités comme des chaînes de longueur un, donc également considérés comme une sous-chaîne.

Pour accéder aux sous-chaînes, utilisez les crochets pour le découpage avec l'index ou les index pour obtenir votre sous-chaîne.

````python
str = 'Hello World!'

print (str[0]) # Imprime le premier caractère de la chaîne
print (str[2:5]) # Imprime les caractères du 3ème au 5ème
print(str[:2]) # Les deux premiers caractères
print(str[2:]) # Tous les caractères sauf les deux premiers
````

Les indices peuvent être négatifs, permettant de compter à partir de la droite.

````python
print(str[-1]) # Le dernier caractère
print(str[-2]) # L'avant dernier caractère
print(str[-2:])# Les deux derniers caractères
print(str[:-2])# Tous les caractères sauf les deux derniers
````

## **Les Listes**

Les listes sont les types de données composés les plus polyvalents de Python. Une liste contient des éléments séparés par des virgules et placés entre crochets ([ ]).

Dans une certaine mesure, les listes sont similaires aux tableaux en C. L'une des différences entre elles est que tous les éléments appartenant à une liste peuvent être de types de données différents.

Les valeurs stockées dans une liste sont accessibles à l'aide de l'opérateur slice ([ ] et [:]) avec des index commençant à 0 au début de la liste et allant jusqu'à la fin -1.

````python
list = [ 'abcd', 786 , 2.23, 'john', 70.2 ]
tinylist = [123, 'john']

print (list)          # Prints complete list
print (list[:])       # Prints complete list
print (list[0])       # Prints first element of the list
print (list[1:3])     # Prints elements starting from 2nd till 3rd
print (list[2:])      # Prints elements starting from 3rd element
print (tinylist * 2)  # Prints list two times
print (list + tinylist) # Prints concatenated lists
````

La fonction intégrée len() s'applique également au listes:

````python
len(list)
````

````python
# Vide la liste: remplace tous les items par une liste vide
list[:] = []
print(list)
````

## **Tuples**

Un tuple est un autre type de données de séquence similaire à la liste. Un tuple se compose d'un certain nombre de valeurs séparées par des virgules. Contrairement aux listes, cependant, les tuples sont placés entre parenthèses.

NB: différence entre les listes et les tuples
- **Les listes sont entre crochets ( [ ] ) et leurs éléments et leur taille peuvent être modifiés**,

- **Tandis que les tuples sont entre parenthèses ( ( ) ) et ne peuvent pas être mis à jour. Les tuples peuvent être considérés comme des listes en lecture seule.**

````python
tuple = ( 'abcd', 786 , 2.23, 'john', 70.2  )
tinytuple = (123, 'john')

print (tuple)           # Prints complete tuple
print (tuple[0])        # Prints first element of the tuple
print (tuple[1:3])      # Prints elements starting from 2nd till 3rd
print (tuple[2:])       # Prints elements starting from 3rd element
print (tinytuple * 2)   # Prints tuple two times
print (tuple + tinytuple) # Prints concatenated tuple
````

Le code suivant n'est pas valide avec le tuple, car nous avons tenté de mettre à jour un tuple, ce qui n'est pas autorisé. Un cas similaire est possible avec les listes


````python

tuple1 = ( 'abcd', 786 , 2.23, 'john', 70.2 )
liste1 = [ 'abcd', 786 , 2.23, 'john', 70.2 ]
print(liste1)
liste1[2] = 1000 # Syntaxe valide avec liste

print(liste1)
````

## **Les dictionnaires**

Les dictionnaires de Python sont en quelque sorte du type table de hachage. Ils fonctionnent comme des tableaux associatifs ou des hachages trouvés dans Perl et se composent de paires clé-valeur. Une clé de dictionnaire peut être presque n'importe quel type Python, mais il s'agit généralement de nombres ou de chaînes. Les valeurs, en revanche, peuvent être n'importe quel objet Python arbitraire.

Les dictionnaires sont entourés d'accolades ({ }) et les valeurs peuvent être attribuées et consultées à l'aide d'accolades ([]).

````python
dict = {}
dict['one'] = "This is one"
dict[2]     = "This is two"

tinydict = {'name': 'john','code':6734, 'dept': 'sales'}

print (dict['one'])       # Prints value for 'one' key
print (dict[2])           # Prints value for 2 key
print (tinydict)          # Prints complete dictionary
print (tinydict.keys())   # Prints all the keys
print (tinydict.values()) # Prints all the values
````

### **Else pour While et For**

Python prend en charge l'association d'une instruction else à une instruction de boucle.




#### Utilisation de l'instruction else avec la boucle For

Si l'instruction else est utilisée avec une boucle for, le bloc else n'est exécuté que si les boucles for se terminent normalement (et non en rencontrant l'instruction break).



````python
numbers = [11,33,55,39,55,75,37,21,23,41,13]

for num in numbers:
   if num%2 == 0:
      print ('the list contains an even number')
      break
else:
   print ('the list doesnot contain even number')
````

#### Utilisation de l'instruction else avec la boucle While

Si l'instruction else est utilisée avec une boucle while, l'instruction else est exécutée lorsque la condition devient fausse.

````python
count = 0
while count < 5:
   print (count, " is  less than 5")
   count = count + 1
else:
   print (count, " is not less than 5")
````

**Arguments par défaut**

Ceci permet de créer une fonction qui peut être appelée avec moins d'arguments.

````python
def ask_ok(prompt, retries=4, complaint='Yes or no, please!'):
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise IOError('refusenik user')
        print(complaint)
````

````python
#en spécifiant seulement l'argument obligatoire:
ask_ok('Do you really want to quit?')
````

````python
# en spécifiant un des arguments optionels:
ask_ok('OK to overwrite the file?', 2)
````

````python
# en spécifiant tous les arguments:
ask_ok('OK to overwrite the file?', 2, 'Come on, only yes or no!')
````

**Arguments de longueur variable**

Vous devrez peut-être traiter une fonction pour plus d'arguments que vous n'en avez spécifiés lors de la définition de la fonction.

Ces arguments sont appelés arguments de longueur variable et ne sont pas nommés dans la définition de la fonction, contrairement aux arguments obligatoires et par défaut.

````python
# Function definition is here
def printinfo( arg1, *vartuple ):
   "This prints a variable passed arguments"
   print ("Output is: ")
   print (arg1)

   for var in vartuple:
      print (var)
   return

# Now you can call printinfo function
printinfo( 10 )
printinfo( 70, 60, 50 )
````

**Les fonctions anonymes**

Ces fonctions sont dites anonymes car elles ne sont pas déclarées de manière standard à l'aide du mot-clé def. Vous pouvez utiliser le mot-clé lambda pour créer de petites fonctions anonymes.

Les formulaires Lambda peuvent prendre n'importe quel nombre d'arguments mais ne renvoient qu'une seule valeur sous la forme d'une expression. Ils ne peuvent pas contenir de commandes ou d'expressions multiples.

Une fonction anonyme ne peut pas être un appel direct à print car lambda nécessite une expression.

Les fonctions Lambda ont leur propre espace de noms local et ne peuvent pas accéder à des variables autres que celles de leur liste de paramètres et celles de l'espace de noms global.


**Pourquoi utiliser les fonctions Lambda ?**

La puissance de lambda est mieux mise en évidence lorsque vous les utilisez en tant que fonction anonyme dans une autre fonction.

Supposons que vous ayez une définition de fonction qui prend un argument, et cet argument sera multiplié par un nombre inconnu :

````python
(lambda x, y: x + y)(2, 3)
````

````python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
````

````python
# Function definition is here
sum = lambda arg1, arg2: arg1 + arg2

# Now you can call sum as a function
print ("Value of total : ", sum( 10, 20 ))
print ("Value of total : ", sum( 87, 35 ))
````

````python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
````
